```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib widget
import pandas as pd
from sklearn import metrics
import seaborn as sn
```

```python
Houses = pd.read_csv('USA_Housing.csv')
```

```python
Houses.head()
```

```python
Houses.info()
```

```python
Houses.describe()
```

```python
Houses.columns
```

```python
x = Houses[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = Houses['Price']
```

```python
x_train, x_test , y_train , y_test = train_test_split(x , y , test_size= 0.4 , random_state= 101)
```

```python
Model = LinearRegression()
```

```python
Model.fit(x_train , y_train)
```

```python
prediction= Model.predict(x_test)
```

```python
plt.scatter(y_test, prediction)
```

```python
MSE_Train = metrics.mean_squared_error(y_test, prediction)
print("The MSE: ", MSE_Train)
```

```python
sn.displot((y_test , prediction))
```
